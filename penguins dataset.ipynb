```python
penguins ["sex"].isnull()
```

```python
penguins ["sex"].isnull().sum
```

```python
penguins['sex'] = penguins['sex'].fillna(penguins['sex'].mode()[0])
```

```python
penguins ["sex"]
```

```python
penguins ["sex"].isnull().sum()
```

```python
penguins ["bill_length_mm"].isnull()
```

```python
penguins ["bill_length_mm"].isnull().sum()
```

```python
penguins ["bill_length_mm"]=penguins["bill_length_mm"].fillna(penguins["bill_length_mm"].median(0))
```

```python
penguins ["bill_length_mm"].isnull().sum()
```

```python
penguins["bill_depth_mm"] = penguins ["bill_depth_mm"].fillna(penguins["bill_depth_mm"].mean(0))
```

```python
penguins["bill_depth_mm"].isnull().sum()
```

```python
penguins["flipper_length_mm"] = penguins ["flipper_length_mm"].fillna(penguins["flipper_length_mm"].mean(0))
```

```python
penguins["body_mass_g"] = penguins ["body_mass_g"].fillna(penguins["body_mass_g"].mean(0))
```

```python
penguins.isnull().sum()
```

```python
penguins['flipper_length_mm'].hist(bins=15, color='skyblue', edgecolor='black')
plt.title('Flipper Length Distribution')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='species', y='bill_length_mm', data=penguins, palette='Set3')
plt.title('Bill Length by Species')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', style='sex', s=80)
plt.title('Bill Length vs Bill Depth')
plt.show()
```

```python
sns.pairplot(penguins, hue='species', corner=True, palette='bright')
plt.show()
```
